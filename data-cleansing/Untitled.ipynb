```python
import pandas as pd
import numpy as np
import os
```

```python
path = os.path.abspath(os.path.join(os.getcwd(), "..\\web-scraping\\trojmiasto-data"))
primary = pd.read_csv(path + "\\data-1-2020-11-26.csv", encoding="cp1250")
secondary = pd.read_csv(path + "\\data-2-2020-11-26.csv", encoding="cp1250")
```

```python
primary["market"] = "pierwotny"
secondary["market"] = "wtórny"
```

```python
df = pd.concat([primary, secondary], ignore_index=True)
```

```python
df.head(5)
```

```python
df.info()
```

```python
df['price'] = df['price'].str.replace(' zł','').str.replace(' ', '').str.replace(',', '.')
df['sq_meter_price'] = df['sq_meter_price'].str.replace(' zł','').str.replace(' ', '').str.replace(',', '.')
df['area'] = df['area'].str.replace(' ', '').str.replace(',', '.')
```

```python
#df['n_rooms'] = df['n_rooms'].astype(np.int64)
#df['build_year'] = df['build_year'].astype(np.int64)
```

```python
# dropping null value columns to avoid errors 
# data.dropna(inplace = True) 
  
# new data frame with split value columns 
new = df["address"].str.split(",", n = 1, expand = True) 
  
# making separate first name column from new data frame 
df["city"]= new[0] 
  
# making separate last name column from new data frame 
df["street"]= new[1] 
  
# Dropping old Name columns 
df.drop(columns =["address"], inplace = True) 
  
# df display 
df 
```

```python
# sort columns
df = df[['company', 'type', 'city', 'street', 'price', 'sq_meter_price', 'area', 'n_rooms', 'floor', 'build_year', 'market']]
```
